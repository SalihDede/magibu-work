```python
import sys, torch
sys.path.insert(0, "deepseek3")                      # match the architecture of the checkpoint
from model import TinyDeepSeek
from tokenizer import CharTokenizer
```

```python

ckpt = torch.load("deepseek3/tiny_deepseek.pt", map_location="cpu", weights_only=False)
tok = CharTokenizer(ckpt["chars"])
model = TinyDeepSeek(ckpt["cfg"]); model.load_state_dict(ckpt["model"]); model.eval()

model
```

```python
start = torch.full((10, 1), tok.eos_id, dtype=torch.long)   # every name starts at EOS/newline
out = model.generate(start, max_new_tokens=model.cfg.max_seq_len,
                     temperature=0.8, eos_id=tok.eos_id)     # stops each row at EOS
out
```

```python
for row in out.tolist():
    print(tok.decode(row[1:]).split("\n")[0])
```
